```python
#EXERCICIO 1 - CALCULAR A MEDIA DE 4 VALORES

#SOLICITAR OS 4 VALORES AO USUARIO - usando float pq depois vamos ter que dividir
valor1 = float(input("Informe o primeiro valor: "))
valor2 = float(input("Informe o segundo valor:  "))
valor3 = float(input("Informe o terceiro valor: "))
valor4 = float(input("Informe o quarto valor:   "))

#SOMAR TODOS OS 4 VALORES UTILIZANDO A VARIAVEL SOMA
soma = valor1 + valor2 + valor3 + valor4

#DEPOIS DE SOMAR OS VALORES VAMOS FAZER A MEDIA QUE VAI SER DIVIDIDA POR 4
media = soma / 4

#NO FINAL EXIBIMOS O RESULTADO DA MEDIA DOS 4 NUMEROS COM DUAS CASAS DECIMAIS
print(f"A media final da soma dos 4 valores e: {media: .2f}")
```

```python
#EXERCICIO 2 - GERAR A TABUADA DO 1 AO 10

#LOOP EXTERNO QUE PERCORRE OS NÚMEROS DE 1 A 10 (QUAL TABUADA SERÁ GERADA)
for i in range(1, 11):

    #EXIBE O TÍTULO DE CADA TABUADA (INDENTADO DENTRO DO LOOP EXTERNO)
    print(f"\n Tabuada do {i}: ")

    #LOOP INTERNO QUE MULTIPLICA O NÚMERO ATUAL DE 1 ATÉ 10
    for j in range(1, 11):

        #CALCULA O RESULTADO DA MULTIPLICAÇÃO (INDENTADO DENTRO DO LOOP INTERNO)
        resultado = i * j

        #EXIBE O RESULTADO DA TABUADA (INDENTADO DENTRO DO LOOP INTERNO)
        print(f"{i} x {j} = {resultado}")
```

```python
#EXERCICIO 4 - MEDIANA USANDO STATISTICS

#IMPORTA A BIBLIOTECA STATISTICS PARA CÁLCULOS ESTATÍSTICOS
import statistics

#DEFINE UMA LISTA COM 5 VALORES NUMÉRICOS
valores = [10, 20, 30, 40, 50]

#CALCULA A MEDIANA DA LISTA (VALOR DO MEIO QUANDO ORDENADO)
mediana = statistics.median(valores)

#EXIBE O RESULTADO DA MEDIANA
print("A mediana é:", mediana)
```

```python
#EXERCICIO 6 - DESVIO PADRÃO USANDO NUMPY

#IMPORTA A BIBLIOTECA NUMPY COM O APELIDO "np"
import numpy as np

#DEFINE UMA LISTA COM 8 VALORES NUMÉRICOS
valores = [10, 12, 23, 23, 16, 23, 21, 16]

#CONVERTE A LISTA PARA UM ARRAY NUMPY (NECESSÁRIO PARA USAR FUNÇÕES DO NUMPY)
array_valores = np.array(valores)

#CALCULA O DESVIO PADRÃO DO ARRAY
desvio = np.std(array_valores)

#EXIBE O RESULTADO DO DESVIO PADRÃO
print("O desvio padrão é:", desvio)
```

```python
#EXERCICIO 7 - CÁLCULO DA AMPLITUDE

#DEFINE UMA LISTA COM 5 VALORES NUMÉRICOS
dados = [8, 12, 15, 20, 25]

#CALCULA A AMPLITUDE: DIFERENÇA ENTRE O MAIOR E O MENOR VALOR
amplitude = max(dados) - min(dados)

#EXIBE O MAIOR VALOR DA LISTA USANDO A FUNÇÃO max()
print("Maior valor:", max(dados))

#EXIBE O MENOR VALOR DA LISTA USANDO A FUNÇÃO min()
print("Menor valor:", min(dados))

#EXIBE O RESULTADO DA AMPLITUDE
print("Amplitude:", amplitude)
```

```python
#EXERCICIO 8 - ANALISE PREDITIVA DE TEMPERATURA

#IMPORTA A BIBLIOTECA NUMPY COM O APELIDO "np"
import numpy as np

#DEFINE UMA LISTA COM 10 REGISTROS DE TEMPERATURA
temperaturas = [22, 23, 21, 24, 25, 23, 22, 26, 27, 24]

#CONVERTE A LISTA PARA UM ARRAY NUMPY
temp_array = np.array(temperaturas)

#CALCULA A MÉDIA DAS TEMPERATURAS
media = np.mean(temp_array)

#CALCULA O DESVIO PADRÃO DAS TEMPERATURAS
desvio = np.std(temp_array)

#CALCULA A AMPLITUDE: DIFERENÇA ENTRE A MAIOR E A MENOR TEMPERATURA
amplitude = np.max(temp_array) - np.min(temp_array)

#EXIBE OS RESULTADOS CALCULADOS
print("Média da temperatura:", media)
print("Desvio padrão:", desvio)
print("Amplitude:", amplitude)

#VERIFICA SE O DESVIO PADRÃO É MAIOR QUE 2 (GRANDE VARIAÇÃO)
if desvio > 2:
    #SE SIM: ALERTA QUE HÁ VARIAÇÃO ALTA E SUGERE VERIFICAÇÃO DO EQUIPAMENTO
    print("Existe grande variação de temperatura. Verificar funcionamento do ar condicionado.")
else:
    #SE NÃO: INFORMA QUE A TEMPERATURA ESTÁ ESTÁVEL
    print("Temperatura está estável. Funcionamento adequado.")
```
